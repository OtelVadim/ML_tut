```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
```

```python
#Generating a dataset :
# y = A*sin(B(X + C)) + D
# A = Amplitude
# Period = 2*pi/B
# Period = Lenght of One Cycle
# C = phase shift (in Radian)
# D = Vertical Shift

x = np.linspace(0,1,100) #(start, end, points)

#here...
#A = 1
#B = 2*pi
#B = 2*pi/period
#period = 1
#C = 0
#D = 0

y = 1*np.sin(2*np.pi*x)

# adding noise:
noise = 0.4 * np.random.normal(size = 100)

y_data = y + noise

plt.scatter(x,y_data,c='red')
```

```python
#calculate the value :
def calc_sine(x,a,b,c,d):
    return a * np.sin(b * (x + np.radians(c))) + d
```

```python
#Finding optimal parameters
popt, pcov = curve_fit(calc_sine, x, y_data)

#plot the main data:
plt.scatter(x,y_data)

#plot the best fit curve:
plt.plot(x,calc_sine(x,*popt), c='r')

#check accuracy:
r2 = r2_score(y_data,calc_sine(x,*popt))
print(r2)
```

```python
#compare with linear regression

def calc_line(X,m,b):
    return b + m*X

#It returns optimized parameters for our function:
#popt stores optimal parameters
#pcov stores the covariance between each parameters.
popt, pcov = curve_fit(calc_line,x,y_data)

#Plot themain data:
plt.scatter(x,y_data)

#plot the best fit line
plt.plot(x,calc_line(x,*popt), c='r')

#check accuracy of linear model:
r2 = r2_score(y_data,calc_line(x,*popt))
print('LinReg accuracy = ', r2)
```
